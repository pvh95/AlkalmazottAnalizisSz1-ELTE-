<b>2.feladat.</b> Írjunk olyan programot, amely előállítja a Newton-féle interpolációs eljáráshoz tanult osztott véges differenciahányodost mátrixos alakban!<br><br>

Alkalmazzuk a programunkat az $$x_1=-1,\ f(x_1)=1,\ x_2=1,\ f(x_2)=-1,\ x_3=2,\ f(x_3)=13,\ x_4=3,\ f(x_4)=69,\ x_5=4,\ f(x_5)=221$$

feladatra.

```octave
% A Newton eljarasban az osztott differencia tablazatra irunk programot
% (f_{i+1}-f_(i))/(x_(i+1)-x_i)

%x = [-1 1 2 3 4];
%f_x = [1 -1 13 69 221];
x = [-1 0 1]
f_x = [1 -1 -1]

n = length(x);
N = zeros(n,n);

% Elso osztott differencia hanyados
for k = 1 : n-1
    N(k,1) = (f_x(k+1) - f_x(k))/(x(k+1)-x(k));
end
% A tobbi osztott differencia hanyados

for j = 2 : n-1
    for k = 1 : n-j
        N(k,j) = (N(k+1, j-1) - N(k, j-1)) / (x(k+j)-x(k));
    end
end

N
Newton = [x', f_x', N]
```

<b>3.feladat.</b> Reprodukáljuk ez az előző 2 feladat eredményét a polyfit beépített függvény segítségével!

```octave
%x=[-1 1 2 3 4];
%f_x=[1 -1 13 69 221];

x = [1 2 3 4];
f_x = [2 1 4 3];

p = polyfit(x,f_x,length(x)-1); % polyfit polinomot illeszt az adatokra
% -1.3333 * x^3 + 10 * x^2 - 21.6667 * x + 15

% xx = x(1):0.01:x(end) % h = (b-a)/N
xx = linspace(x(1),x(end),100); %N

Lagrange = polyval(p, xx); %polinom kiertekeles polyval(egyutthato vektor, alappontok)

plot(x,f_x,'mo',xx,Lagrange)
legend('Adatok','Lagrange-fele intpol. pol.')
```
