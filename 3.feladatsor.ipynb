1.feladat. Tekintsük az $n$ alappontra támaszkodó Lagrange-féle interpoláló polinomot

$$ L_n(x)=\sum_{i=1}^nf(x_i)\prod_{\substack{j=1 \\ j\neq i}}^n\frac{x-x_j}{x_i-x_j} $$

és ahhoz tartozó Dan Ellis által (Columbia Univerity, USA) megírt lagrangepoly.m fájlt. Értelmezzük a kódot és alkalmazzuk az $$x_1=1,\ f(x_1)=2,\ x_2=2,\ f(x_2)=1,\ x_3=3,\ f(x_3)=4,\ x_4=4,\ f(x_4)=3$$

feladatra. Melyik interpoláló polinomot kaptuk? Készítsünk eredményükről a lagrangepoly.m fájl és a polyval beépített parancs segítségével informatív ábrát!

```octave
X = [1 2 3 4];
Y = [2 1 4 3];
lagrangepoly(X,Y)
x = 1 : 0.01 : 4;
%plot(x,polyval(lagrangepoly(X,Y),x),X,Y,'or',R,S,'.b')
plot(x,polyval(lagrangepoly(X,Y),x),X,Y,'or')
title('Lagrange-fele interpolacios polinom')
```

2.feladat. Írjunk olyan programot, amely előállítja a Newton-féle interpolációs eljáráshoz tanult osztott véges differenciahányodost mátrixos alakban!<br><br>

Alkalmazzuk a programunkat az $$x_1=-1,\ f(x_1)=1,\ x_2=1,\ f(x_2)=-1,\ x_3=2,\ f(x_3)=13,\ x_4=3,\ f(x_4)=69,\ x_5=4,\ f(x_5)=221$$

feladatra.

```octave
% Newton-féle interpolációhoz tartozo mátrix
x=[-1 1 2 3 4];
f_x=[1 -1 13 69 221];
n = length(x);
% Elso osztott veges differencia hanyados
for k = 1 : n - 1
   N(k, 1) = (f_x(k+1) - f_x(k))/(x(k+1) - x(k));
end
% A tobbi osztott veges differencia hanyados
for j = 2 : n - 1
   for k = 1 : n - j
      N(k, j) = (N(k+1, j - 1) - N(k, j - 1))/(x(k+j) - x(k));
   end
end
N
```

```octave
% Newton-féle interpolációhoz tartozo mátrix
x=[-1 1 2 3 4];
f_x=[1 -1 13 69 221];
n = length(x);
N=zeros(n-1);
% Elso osztott veges differencia hanyados
for k = 1 : n - 1
   N(k, 1) = (f_x(k+1) - f_x(k))/(x(k+1) - x(k));
end
N
```
